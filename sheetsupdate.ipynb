```python
drop_number = sheet.worksheet_by_title('Daily drop numbers')
drop_number.get_as_df()
```

```python
drop_number_df = drop_number.get_as_df()
drop_number_df['Date'] = pd.to_datetime(drop_number_df['Date'])
drop_number_df.set_index('Date', inplace=True)

last_date = drop_number_df.index[-1].to_pydatetime().date()    
today = pd.Timestamp.today().date()
playercount = scraper.get_playercount()
daily_peak = playercount.get('24-hour peak',0)
playercount = playercount.get('Playercount',0)
if playercount > daily_peak:
    daily_peak = playercount
if  today != last_date:
    index = drop_number_df.reset_index().index[-1]
    date_cell = drop_number.cell(f'A{index+3}')
    date_cell.set_value(str(pd.Timestamp.today().date()))
    playercount_cell = drop_number.cell(f'B{index+3}')
    playercount_cell.set_value(daily_peak)
# Handles situations when there is a more accurate 24-hour peak available
elif daily_peak != drop_number_df.Playercount[-1]:
    index = drop_number_df.reset_index().index[-1]
    date_cell = drop_number.cell(f'A{index+2}')
    date_cell.set_value(str(pd.Timestamp.today().date()))
    playercount_cell = drop_number.cell(f'B{index+2}')
    playercount_cell.set_value(daily_peak)
```
